```python
import requests
import bs4
import encodings
import pandas
```

```python
def get_url_return_soup_table():
    url = 'https://e-service.cwb.gov.tw/wdps/obs/state.htm#description'
    html = requests.get(url)
    html.encoding = "utf-8"
    soup = bs4.BeautifulSoup(html.text , 'html.parser')
    return soup.table
```

```python
def split_list(datalist, n):
  # 將list分割 (l:list, n:每個matrix裡面有n個元素)
  for idx in range(0, len(datalist), n):
    yield datalist[idx:idx+n]
```

```python
def output_data():
#main code
    now_exist_station_table = get_url_return_soup_table()
    
#find columns list
    search_columns = now_exist_station_table.find_all('th')
    columns_list = []
    for element in search_columns:
        columns_list.append(element.text)

#find all station data  
    data = now_exist_station_table.find_all('td')
    data_list = []

    for element in data:
        if element.text == "":
            data_list.append("無")
        else:
            data_list.append(element.text)

#split every station data
    data_list = list( split_list(data_list , 12) )

#create dataframe to input data
    station_form = pandas.DataFrame(list( data_list ),columns = columns_list)
                                
    station_form.index+=1

#output data by excel
    return station_form
```

```python
station_data = output_data()
station_data
```
